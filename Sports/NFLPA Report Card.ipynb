```python
grade_mapping = {
    'A+': 100,
    'A': 96,
    'A-': 92,
    'B+': 89,
    'B': 86,
    'B-': 82,
    'C+': 79,
    'C': 76,
    'C-': 72,
    'D+': 69,
    'D': 66,
    'D-': 62,
    'F': 59,
    'F-': 55
}

order = {
    'A+': 1,
    'A': 2,
    'A-': 3,
    'B+': 4,
    'B': 5,
    'B-': 6,
    'C+': 7,
    'C': 8,
    'C-': 9,
    'D+': 10,
    'D': 11,
    'D-': 12,
    'F': 13,
    'F-': 14
}
```

## 2023 Season

```python
df = pd.read_excel('Combined NFLPA Report Card.xlsx', sheet_name='2023')

grade_columns = ['Treatment of Families',
                 'Nutrition',
                 'Weight Room',
                 'Strength Staff',
                 'Training Room',
                 'Training Staff',
                 'Locker Room',
                 'Travel']

for col in grade_columns:
    df[f'{col} 2023'] = df[col].map(grade_mapping)

numeric_columns = ['Treatment of Families 2023',
                 'Nutrition 2023',
                 'Weight Room 2023',
                 'Strength Staff 2023',
                 'Training Room 2023',
                 'Training Staff 2023',
                 'Locker Room 2023',
                 'Travel 2023']

df['2023 Averages'] = df[numeric_columns].mean(axis=1)

df_sorted = df.sort_values(by='2023 Averages', ascending=False)

print(df_sorted.round(2))
```

### User Interface for Selecting Desired Season, Category, Grade

```python
import tkinter as tk
from tkinter import ttk, messagebox

excel_file = 'Combined NFLPA Report Card.xlsx'  # Make sure to use the correct path
sheet_data = pd.read_excel(excel_file, sheet_name=None)

# Assuming 'sheet_data' contains the processed DataFrames for each sheet (already imported)

# Step 3: Function to search and display results
def search_and_display():
    sheet_name = sheet_combobox.get()
    row_value = row_combobox.get()
    column_value = column_combobox.get()

    if not sheet_name or not row_value or not column_value:
        messagebox.showerror("Error", "Please select all fields.")
        return

    df = sheet_data[sheet_name]

    # Check if the row value exists in the dataframe
    try:
        row_index = df[df[column_value] == row_value].index[0]
        result_value = df.at[row_index, column_value]
        result_label.config(text=f"Found value: {result_value}")
    except IndexError:
        messagebox.showerror("Error", f"Value '{row_value}' not found in column '{column_value}'.")
    
# Step 4: Create the Tkinter UI
def create_ui():
    # Create the main window
    root = tk.Tk()
    root.title("Excel Search Interface")
    
    # Label and dropdown for sheet selection
    tk.Label(root, text="Select Sheet:").grid(row=0, column=0, padx=10, pady=10)
    global sheet_combobox
    sheet_combobox = ttk.Combobox(root, values=list(sheet_data.keys()), state="readonly")
    sheet_combobox.grid(row=0, column=1, padx=10, pady=10)

    # Label and dropdown for row selection
    tk.Label(root, text="Enter Row Value:").grid(row=1, column=0, padx=10, pady=10)
    global row_combobox
    row_combobox = ttk.Combobox(root, state="readonly")
    row_combobox.grid(row=1, column=1, padx=10, pady=10)

    # Label and dropdown for column selection
    tk.Label(root, text="Select Column:").grid(row=2, column=0, padx=10, pady=10)
    global column_combobox
    column_combobox = ttk.Combobox(root, state="readonly")
    column_combobox.grid(row=2, column=1, padx=10, pady=10)

    # Label to display the result
    global result_label
    result_label = tk.Label(root, text="Result will be shown here.")
    result_label.grid(row=3, column=0, columnspan=2, padx=10, pady=10)

    # Search Button
    search_button = tk.Button(root, text="Search", command=search_and_display)
    search_button.grid(row=4, column=0, columnspan=2, pady=20)

    # Update the row and column dropdown options based on sheet selection
    def update_dropdowns(event):
        sheet_name = sheet_combobox.get()
        if sheet_name:
            # Update the column and row comboboxes
            df = sheet_data[sheet_name]
            row_combobox['values'] = df[column_combobox.get()].unique().tolist()  # Row values based on column
            column_combobox['values'] = df.columns.tolist()  # Column names
            column_combobox.set(df.columns[0])  # Set default column

    # Bind sheet selection to update dropdowns
    sheet_combobox.bind("<<ComboboxSelected>>", update_dropdowns)
    
    # Run the UI
    root.mainloop()

# Create the user interface
create_ui()
```

```python
# Grade map for letter grades to numeric values
def grade_to_numeric(grade):
    grade_map = {
    'A+': 100,
    'A': 96,
    'A-': 92,
    'B+': 89,
    'B': 86,
    'B-': 82,
    'C+': 79,
    'C': 76,
    'C-': 72,
    'D+': 69,
    'D': 66,
    'D-': 62,
    'F': 59,
    'F-': 55
}
    return grade_map.get(grade.upper(), None)  # Return None if not a valid grade

# Define weights for each assignment (make sure these add up to 1.0 or 100%)
weights = {
    'Treatment of Families': 0.06,
    'Nutrition': 0.06,
    'Weight Room': 0.09,
    'Strength Staff': 0.09,
    'Training Room': 0.09,
    'Training Staff': 0.09,
    'Locker Room': 0.09,
    'Travel': 0.09,
    'Food/Cafeteria': 0.09,
    'Head Coach': 0.1,
    'Owner': 0.15
}
```

```python
# Convert all letter grades to numeric values
for col in df.columns:
    df[col] = df[col].apply(grade_to_numeric)
```

```python
selection = input("Choose either 'grade' or 'numeric':")
if selection == "grade":
    column_name = input("Select one of the following: Weight Room, Strength Staff, Training Room, Training Staff, Nutrition, Travel, Head Coach, Owner, 2023 Averages, 2024 Averages, Treatment of Families, Food/Cafeteria")
    grade = input("Select a letter grade (can include + or -)")

    df_filtered = df[df[column_name].str.contains(grade, na=False)][['Team', column_name]]

    df_filtered['Order Rank'] = df_filtered[column_name].map(order)
    df_filtered_sorted = df_filtered.sort_values(by="Order Rank").drop(columns=["Order Rank"])

    print(df_filtered_sorted)

elif selected == "numeric":
    column_name = input("Select one of the following: Weight Room, Strength Staff, Training Room, Training Staff, Nutrition, Travel, Head Coach, Owner, 2023 Averages, 2024 Averages, Treatment of Families, Food/Cafeteria")
    grade = input("Select a numeric value")
    try:
        grade = float(grade) if '.' in grade else int(grade)
        is_numeric = True
    except ValueError:
        is_numeric = False

    if is_numeric:
        df_filtered = df[df[column_name].apply(lambda x: isinstance(x, (int, float)) and x > grade)][['Team', column_name]]
    else: 
        df_filtered = df[df[column_name].astype(str).str.contains(grade, na=False)][['Team', column_name]]

    if not df_filtered.empty:
        if df_filtered[column_name].dtype == "O":
            df_filtered["Order_Rank"] = df_filtered[column_name].map(order)
            df_filtered_sorted = df_filtered.sort_values(by="Order_Rank", na_position="last").drop(columns=["Order Rank"])
        else:
            df_filtered_sorted = df_filtered

else:
    print("Please try again")

print(df_filtered_sorted)
        
```
