```python
import csv
import os
```

```python
FILENAME = "indycar_driver_data.csv"

# Ensure the file has headers if it doesn't exist
if not os.path.isfile(FILENAME):
    with open(FILENAME, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([
            "Race", "Driver", "Starting Position", "Finishing Position",
            "Laps Led", "Points Earned"
        ])

def add_driver_data():
    race = input("Enter race name or number: ")
    driver = input("Enter driver name: ")
    start_pos = input("Enter starting position: ")
    finish_pos = input("Enter finishing position: ")
    laps_led = input("Enter number of laps led: ")
    points = input("Enter points earned: ")

    with open(FILENAME, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([race, driver, start_pos, finish_pos, laps_led, points])
    print(f"Data saved for {driver} in race {race}.")

def main():
    while True:
        add_driver_data()
        another = input("Add another? (y/n): ").lower()
        if another != 'y':
            break

if __name__ == "__main__":
    main()
```
