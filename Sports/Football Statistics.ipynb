Using the table to determine APY/catch

```python
data = {
    'Player': ['Likely', 'Flowers', 'Bateman', 'Algholor', 'Andrews', 'Kolar'], 
    'APY': [4164336, 3509110, 6435949, 3750000, 14000000, 1091220],
    'Catches': [13, 17, 9, 5, 6, 2] 
}
df = pd.DataFrame(data)
print(df)
```

```python
df[['price_per_catch']] = df[['APY']].div(df['Catches'], axis=0)
print(df)
```
