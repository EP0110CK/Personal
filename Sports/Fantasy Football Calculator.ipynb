```python
print("Player Position:")
flex = ["Running Back","running back","Running back","RB","rb","Wide Receiver","wide receiver","Wide receiver","WR","wr","Tight End", "Tight end", "tight end", "TE", "te"]
qb = ["Quarterback","quarterback","QB","qb"]
dst = ['D/ST', 'd/st', 'Defense', 'defense', 'DST', 'dst']
k = ['Kicker', 'kicker', 'K', 'k']

position = input()
while position != "":
    if position in qb:
        pass_yards = input("Passing Yards:")
        pass_tds = input("Passing Touchdowns:")
        pass_ints = input("Passing Interceptions:")
        rushing_yards = input("Rushing Yards:")
        rushing_touchdown = input("Rushing Touchdowns:")
        fumble = input("Number of Fumbles:")
        print("---------------------------------")
        print("Passing Yards:", pass_yards)
        print("Passing Touchdowns:", pass_tds)
        print("Passing Interceptions:", pass_ints)
        print("-------------------------------------")
        print("Rushing Yards:", rushing_yards)
        print("Rushing Touchdowns:", rushing_touchdown)
        print("Fumbles:", fumble)

        fantasy_pass_yards = int(pass_yards)*0.04
        fantasy_pass_tds = int(pass_tds)*4
        fantasy_pass_ints = int(pass_ints)*2
        fantasy_rush_yards = int(rushing_yards)*0.1
        fantasy_rush_touchdown = int(rushing_touchdown)*6
        fantasy_fumble = int(fumble)*2
        qb_fantasy_points = fantasy_pass_yards + fantasy_pass_tds - fantasy_pass_ints + fantasy_rush_yards + fantasy_rush_touchdown - fantasy_fumble
        print("---------------------------------")
        print("Quarterback Fantasy Points:", round(qb_fantasy_points, 2))
        break
    elif position in flex:
        rushing_yards = input("Rushing Yards:")
        rushing_touchdown = input("Rushing Touchdowns:")
        fumble = input("Number of Fumbles:")
        receptions = input("Receptions:")
        receiving_yards = input("Receiving Yards:")
        reception_touchdown = input("Receiving Touchdowns:")
        print("-------------------------------------")
        print("Rushing Yards:", rushing_yards)
        print("Rushing Touchdowns:", rushing_touchdown)
        print("Fumbles:", fumble)
        print("-------------------------------------")
        print("Receptions:", receptions)
        print("Receiving Yards:", receiving_yards, "yards")
        print("Touchdown Receptions:", reception_touchdown)
        
        fantasy_rush_yards = int(rushing_yards)*0.1
        fantasy_rush_touchdown = int(rushing_touchdown)*6
        fantasy_fumble = int(fumble)*2
        fantasy_reception = int(receptions)
        fantasy_receiving_yards = int(receiving_yards)*0.1
        fantasy_receiving_touchdown = int(reception_touchdown)*6
        flex_fantasy_points = fantasy_rush_yards + fantasy_rush_touchdown - fantasy_fumble + fantasy_reception + fantasy_receiving_yards + fantasy_receiving_touchdown
        print("Total Fantasy Points:", round(flex_fantasy_points, 2))
        break
    elif position in dst:
        yards_allowed = input("Yards Allowed:")
        points_allowed = input("Points Allowed:")
        ints = input("Interceptions:")
        sacks = input("Sacks:")
        fumbles = input("Fumble Recoveries:")
        dst_td = input("Defensive/Special Teams Touchdowns:")
        
        break
    elif position in k:
        xp_good = input("Extra Points Made:")
        xp_attempt = input("Extra Points Attempted:")
        
        break
    else:
        position = input("Try Another Position:")
```

```python
def quarterback(yards, td, int, rush_yds, rush_td, fum):
    yards_pts = yards * 0.04
    td_pts = td * 4
    int_pts = int * 2
    rush_yds_pts = rush_yds * 0.1
    rush_td_pts = rush_td * 6
    fum_pts = fum * 2
    points = yards_pts + td_pts - int_pts + rush_yds_pts + rush_td_pts - fum_pts
    print("Fantasy Points:", points)
    
quarterback(input(Passing Yards:), input(Passing Touchdowns:), input(Interceptions Thrown:), input(Rushing Yards:), input(Rushing Touchdowns), input(Fumbles Lost))
print(quarterback)
```
