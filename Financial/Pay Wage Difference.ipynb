```python
hourly_wage = input("Hourly Pay Rate: $")
hourly_hours = input("Hours Worked Per Week: $")
gross_pay = float(hourly_wage) * float(hourly_hours)
net_pay = gross_pay * 0.75
print("---------------------------------------------")
print("Monthly Gross Income: $", round((gross_pay*4), 2))
print("Monthly Net Income: $", round((net_pay*4), 2))
```

```python
intern_wage = input("Hourly Intern Rate: $")
intern_hours = input("Hours Worked Per Week:")
if float(intern_hours) > 40:
    intern_overtime = float(intern_wage) * 1.5
    hours_overtime = (float(intern_wage) * 40) + (float(intern_overtime) * (float(intern_hours) - 40))
    net_overtime = hours_overtime * 0.75
    print("---------------------------------------------")
    print("Monthly Gross Income: $", round((hours_overtime*4), 2))
    print("Monthly Net Income: $", round((net_overtime*4), 2))
else:   
    gross_intern = float(intern_wage) * float(intern_hours)
    net_intern = gross_intern * 0.75
    print("---------------------------------------------")
    print("Monthly Gross Income: $", round((gross_intern*4), 2))
    print("Monthly Net Income: $", round((net_intern*4), 2))
```

```python
difference = (net_pay - (net_intern or net_overtime))
print("Difference in monthly net income is: $", round((difference*4), 2))
```
