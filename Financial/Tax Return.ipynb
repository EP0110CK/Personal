```python
choice = input("Insert whether you have 'final' annual income or want to 'calculate' final annual income:")

if choice == 'calculate':
    annual_so_far = input("Insert total yearly income so far: $")
    last_paycheck = input("Insert last paycheck amount (gross amount): $")
    annual_fed_so_far = input("Insert total federal taxes withheld so far: $")
    annual_state_so_far = input("Insert total state taxes withheld so far: $")
    annual_local_so_far = input("Insert total local taxes withheld so far: $")
    annual_remaining_paychecks = input("Enter how many more pay periods in the calendar year: ")
    fed_WH_rate = float(annual_fed_so_far)/float(annual_so_far)
    state_WH_rate = float(annual_state_so_far)/float(annual_so_far)
    local_WH_rate = float(annual_local_so_far)/float(annual_so_far)
    estimated_income = float(annual_so_far) + (float(last_paycheck) * int(annual_remaining_paychecks))
    gross_income = estimated_income
    estimated_fed_WH = float(annual_fed_so_far) + ((float(estimated_income) * fed_WH_rate) - float(annual_fed_so_far))
    fed_WH = estimated_fed_WH
    estimated_state_WH = float(annual_state_so_far) + ((float(estimated_income) * state_WH_rate) - float(annual_state_so_far))
    state_WH = estimated_state_WH
    estimated_local_WH = float(annual_local_so_far) + ((float(estimated_income) * local_WH_rate) - float(annual_local_so_far))
    local_WH = estimated_local_WH
    print("Estimated yearly income: $", round(float(gross_income), 2))
    print("Estimated federal taxes withheld: $", round(float(fed_WH), 2))
    print("Estimated state taxes withheld: $", round(float(state_WH), 2))
    print("Estimated local taxed withheld: $", round(float(local_WH), 2))

else:
    gross_income = input("Insert total yearly income (box 1):")
    fed_WH = input("Insert federal taxes withheld (box 2):")
    state_WH = input("Insert state taxes withheld (box 17):")
    local_WH = input("Insert local taxes withheld (box 18):")


print("-------------------------------------------------")

std_ded = 14600
taxable_income = int(gross_income) - std_ded
print("Taxable Income: $", taxable_income)

print("-------------------------------------------------")

if taxable_income < 11600:
    tax = taxable_income * 0.1
elif taxable_income <47150:
    tax = 1160 + ((taxable_income - 1160) * 0.12)
elif taxable_income < 100525:
    tax = 5426 + ((taxable_income - 5426) * 0.22)
elif taxable_income < 191950:
    tax = 17168.5 + ((taxable_income - 17168.5) * 0.24)
elif taxable_income < 243725:
    tax = 39110.5 + ((taxable_income - 39110.5) * 0.32)
elif taxable_income < 609350:
    tax = 55678.5 + ((taxable_income - 55678.5) * 0.35)
else:
    tax = 183647.25 + ((taxable_income - 183647.25) * 0.37)
print("Federal Taxes: $", round(float(tax), 2))


if taxable_income < 1000:
    MD = taxable_income * 0.02
elif taxable_income < 2000:
    MD = 20 + ((taxable_income - 20) * 0.03)
elif taxable_income < 3000:
    MD = 50 + ((taxable_income - 50) * 0.04)
elif taxable_income < 100000:
    MD = 90 + ((taxable_income - 90) * 0.0475)
elif taxable_income < 125000:
    MD = 4697.5 + ((taxable_income - 4697.5) * 0.05)
elif taxable_income < 150000:
    tax = 5947.5 + ((taxable_income - 5947.5) * 0.0525)
elif taxable_income < 250000:
    MD = 7260 + ((taxable_income - 7260) * 0.055)
else:
    MD = 12760 + ((taxable_income - 12760) * 0.0575)
print("State Taxes: $", round(float(MD), 2))


HoCo = taxable_income * 0.032
print("Local Taxes: $", round(float(HoCo), 2))





print("--------------------------------------------------")
fed_refund = float(fed_WH) - float(tax)
if fed_refund > 0:
    print("Federal refund amount: $", round(float(fed_refund), 2))
elif fed_refund < 0:
    print("Federal amount owed: $", round(float(fed_refund), 2))
else:
    print("No federal refund or amount owed")
  

state_refund = float(state_WH) - float(MD)
if state_refund > 0:
    print("State refund amount: $", round(float(state_refund), 2))
elif state_refund < 0:
    print("State amount owed: $", round(float(state_refund), 2))
else:
    print("No state refund or amount owed")
   

local_refund = float(local_WH) - float(HoCo)
if state_refund > 0:
    print("Local refund amount: $", round(float(local_refund), 2))
elif local_refund < 0:
    print("Local amount owed: $", round(float(local_refund), 2))
else:
    print("No local refund or amount owed")
```
