Recommended Monthly Budget Breakdown

```python
print("NEEDS")
housing = input("Enter how much you spend on housing each month: $")
transport = input("Enter how much you spend on transportation each month: $")
groceries = input("Enter how much you spend on groceries each month: $")
health = input("Enter how much you spend on healthcare each month: $")
debt = input("Enter how much you spend on debt payments each month: $")

total_needs = int(housing) + int(transport) + int(groceries) + int(health) + int(debt)
print("Total 'Needs' Expenses: $", total_needs)

print("-------------------------------------------------------")
print("WANTS")
entertainment = input("Enter how much you spend on entertainment each month: $")
dining = input("Enter how much you spend on eating out each month: $")
hobbies = input("Enter how much you spend on hobbies each month: $")
splurges = input("Enter how much you splurge with each month: $")

total_wants = int(entertainment) + int(dining) + int(hobbies) + int(splurges)
print("Total 'Wants' Expenses: $", total_wants)

print("-------------------------------------------------------")
print("SAVINGS")
emergency = input("Enter how much you put towards your emergency fund each month: $")
retirement = input("Enter how much you put towards your retirement each month: $")
travel = input("Enter how much you put towards travel and vacations each month: $")

total_saving = int(emergency) + int(retirement) + int(travel)
print("Total 'Savings' Expenses: $", total_saving)
```

```python
total_monthly = total_needs + total_wants + total_saving
print("Total Spent Per Month: $", total_monthly)
needs_percent = round((total_needs/total_monthly)*100, 2)
print(f"NEEDS {needs_percent}%")
wants_percent = round((total_wants/total_monthly)*100, 2)
print(f"WANTS {wants_percent}%")
savings_percent = round((total_saving/total_monthly)*100, 2)
print(f"SAVINGS {savings_percent}%")
```
